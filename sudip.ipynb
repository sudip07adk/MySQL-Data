```python
%load_ext sql
# Connect to SQLite database
%sql sqlite:///simple_database.db
import sqlite3
import pandas as pd

# Create and connect to database
conn = sqlite3.connect('simple_db.db')
cursor = conn.cursor()

# Create table
cursor.execute('''CREATE TABLE IF NOT EXISTS Students (
                    id INTEGER PRIMARY KEY,
                    name TEXT,
                    age INTEGER,
                    grade TEXT
                )''')

# Insert data
students_data = [
    (1, 'Sudip', 20, 'A'),
    (2, 'Ram', 22, 'B'),
    (3, 'Sita', 21, 'A'),
    (4, 'Susmita', 23, 'C')
]

cursor.executemany('INSERT OR IGNORE INTO Students VALUES (?, ?, ?, ?)', students_data)
conn.commit()

# Display data
df = pd.read_sql_query('SELECT * FROM Students', conn)
df
```
